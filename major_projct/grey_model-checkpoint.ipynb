```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
from sklearn.model_selection import train_test_split
```

```python
sar_path = r'C:\Users\spand\OneDrive\Desktop\major_project\dataset\sar'
rgb_path = r'C:\Users\spand\OneDrive\Desktop\major_project\dataset\rgb'

sar_images = sorted(glob(os.path.join(sar_path, '*.png')))
rgb_images = sorted(glob(os.path.join(rgb_path, '*.png')))

print(f'Total SAR images: {len(sar_images)}')
print(f'Total RGB images: {len(rgb_images)}')
```

```python
#to compare sar and rgb from the dataset side by side 
def load_images(index=0):
    sar = cv2.imread(sar_images[index], cv2.IMREAD_GRAYSCALE)
    rgb = cv2.cvtColor(cv2.imread(rgb_images[index]), cv2.COLOR_BGR2RGB)
    return sar, rgb

sar, rgb = load_images(0)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('SAR (Grayscale)')
plt.imshow(sar, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('RGB (Target)')
plt.imshow(rgb)
plt.show()
```

```python
IMG_SIZE = 128

def preprocess(index):
    sar = cv2.resize(cv2.imread(sar_images[index], cv2.IMREAD_GRAYSCALE), (IMG_SIZE, IMG_SIZE))
    rgb = cv2.resize(cv2.cvtColor(cv2.imread(rgb_images[index]), cv2.COLOR_BGR2RGB), (IMG_SIZE, IMG_SIZE))
    
    sar = sar.astype('float32') / 255.0
    rgb = rgb.astype('float32') / 255.0

    sar = np.expand_dims(sar, axis=-1)   # (H, W, 1)
    
    return sar, rgb

# Prepare full dataset
X, Y = [], []
for i in range(len(sar_images)):
    sar, rgb = preprocess(i)
    X.append(sar)
    Y.append(rgb)

X = np.array(X)
Y = np.array(Y)

print("Shape of SAR input:", X.shape)
print("Shape of RGB target:", Y.shape)
```

*U-Net Model for Colorization

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, UpSampling2D, concatenate
from tensorflow.keras.models import Model

def build_unet(input_shape=(128, 128, 1)):
    inputs = Input(input_shape)

    # Encoder
    c1 = Conv2D(64, (3,3), activation='relu', padding='same')(inputs)
    p1 = MaxPooling2D((2,2))(c1)

    c2 = Conv2D(128, (3,3), activation='relu', padding='same')(p1)
    p2 = MaxPooling2D((2,2))(c2)

    c3 = Conv2D(256, (3,3), activation='relu', padding='same')(p2)
    p3 = MaxPooling2D((2,2))(c3)

    # Bottleneck
    bn = Conv2D(512, (3,3), activation='relu', padding='same')(p3)

    # Decoder
    u3 = UpSampling2D((2,2))(bn)
    m3 = concatenate([u3, c3])
    c4 = Conv2D(256, (3,3), activation='relu', padding='same')(m3)

    u2 = UpSampling2D((2,2))(c4)
    m2 = concatenate([u2, c2])
    c5 = Conv2D(128, (3,3), activation='relu', padding='same')(m2)

    u1 = UpSampling2D((2,2))(c5)
    m1 = concatenate([u1, c1])
    c6 = Conv2D(64, (3,3), activation='relu', padding='same')(m1)

    outputs = Conv2D(3, (1,1), activation='sigmoid')(c6)  # 3 channels for RGB

    model = Model(inputs, outputs)
    return model
```

```python
unet = build_unet(input_shape=(128, 128, 1))
unet.compile(optimizer='adam', loss='mse', metrics=['mae'])
unet.summary()
```

```python
from sklearn.model_selection import train_test_split

# Assuming X and Y are your full datasets from preprocessing step
X_train, X_val, Y_train, Y_val = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Training set:", X_train.shape, Y_train.shape)
print("Validation set:", X_val.shape, Y_val.shape)
```

TRAIN THE MODEL

```python
history = unet.fit(X_train, Y_train, validation_data=(X_val, Y_val), 
                   epochs=20, batch_size=8)
```

```python
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
import numpy as np
import matplotlib.pyplot as plt

# Predict on validation data
preds = unet.predict(X_val)

# Initialize accumulators
total_psnr = 0
total_ssim = 0
n = len(X_val)

# Loop through each predicted and actual image pair
for i in range(n):
    pred_img = preds[i]
    true_img = Y_val[i]
    
    # If images are scaled to [0,1], convert to [0,255] and uint8 for SSIM
    pred_img_uint8 = (pred_img * 255).astype(np.uint8)
    true_img_uint8 = (true_img * 255).astype(np.uint8)
    
    # Compute PSNR and SSIM
    psnr_score = psnr(true_img_uint8, pred_img_uint8, data_range=255)
    ssim_score = ssim(true_img_uint8, pred_img_uint8, data_range=255, channel_axis=-1)
    
    total_psnr += psnr_score
    total_ssim += ssim_score

# Average metrics
avg_psnr = total_psnr / n
avg_ssim = total_ssim / n

print(f"Average PSNR: {avg_psnr:.2f} dB")
print(f"Average SSIM: {avg_ssim:.4f}")
```
